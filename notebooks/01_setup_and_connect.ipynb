## Mode A: Connect to an Existing Foundry Project

If you already have an Azure AI Foundry project, paste your connection string below.

**Where to find it**: Azure Portal → AI Foundry → Your Project → Settings → Connection string

````python
# Replace with your actual connection string
CONNECTION_STRING = "<your-connection-string>"
````

````python
from azure.ai.agents import AgentsClient
from azure.identity import DefaultAzureCredential

client = AgentsClient(
    endpoint=CONNECTION_STRING,
    credential=DefaultAzureCredential()
)

print("✓ Connected to Azure AI Foundry project successfully!")
````

## Mode B: Provision a New Foundry Project

If you need to create a new Foundry project from scratch, use the Terraform or Bicep templates:

**Terraform**:
```bash
cd infra/terraform
terraform init
terraform apply -var-file=envs/dev.tfvars -var="use_existing_foundry=false"
```

**Bicep**:
```bash
az deployment sub create \
  --location eastus \
  --template-file infra/bicep/main.bicep \
  --parameters infra/bicep/parameters/dev.bicepparam
```

After provisioning, copy the output connection string and set it in the cell above.

## Verify Connection

List existing agents to verify the connection is working.

````python
agents_list = list(client.list_agents())
print(f"Found {len(agents_list)} existing agent(s):")
for agent in agents_list:
    print(f"  - {agent.name} (id: {agent.id}, model: {agent.model})")
````
